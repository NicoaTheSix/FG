```python
# reduce_and_metrics.py
import sys, os
import json

# 1. 把 backend 加到 path，才能 import utils.pipeline
project_root = r"D:\Log web new\web-project\backend"
if project_root not in sys.path:
    sys.path.insert(0, project_root)

from utils.pipeline import reduce_cpr, reduce_fd

def load_graph(path):
    with open(path, 'r', encoding='utf-8') as f:
        return json.load(f)

def save_graph(graph, path):
    with open(path, 'w', encoding='utf-8') as f:
        json.dump(graph, f, ensure_ascii=False, indent=2)

def count_nodes_edges(graph):
    nodes = set()
    for e in graph['edges']:
        d = e['data']
        nodes.add(d['source'])
        nodes.add(d['target'])
    return len(nodes), len(graph['edges'])

def main():
    src_path = r"D:\Log web new\web-project\backend\static\output\latest_graph.json"
    out_dir  = os.path.dirname(src_path)

    graph0 = load_graph(src_path)
    n0, e0 = count_nodes_edges(graph0)

    graph1 = reduce_cpr(graph0)
    n1, e1 = count_nodes_edges(graph1)
    # save_graph(graph1, os.path.join(out_dir, "latest_graph_cpr.json"))

    graph2 = reduce_fd(graph0)
    n2, e2 = count_nodes_edges(graph2)
    # save_graph(graph2, os.path.join(out_dir, "latest_graph_fd.json"))

    results = [
        {'method': 'original', 'nodes': n0, 'edges': e0},
        {'method': 'CPR',      'nodes': n1, 'edges': e1},
        {'method': 'FD',       'nodes': n2, 'edges': e2},
    ]
    with open(os.path.join(out_dir, "reduction_metrics_C8.json"), 'w', encoding='utf-8') as f:
        json.dump(results, f, ensure_ascii=False, indent=2)

if __name__ == "__main__":
    main()
```
